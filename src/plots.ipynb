```python
import numpy as np
import matplotlib.pyplot as plt

from simulation import (
    monte_carlo,
    summarize_results,
    plot_sampling_distribution,
    plot_confidence_intervals,
    convergence_experiment
)
```

```python
# Hyperparameters
mu = 1.0
sigma = 2.0 # Standard deviation of noise
alpha = 0.05
n = 30
n_rep = 1000
```

```python
results = monte_carlo(
    n=n,
    mu=mu,
    sigma=sigma,
    alpha=alpha,
    n_rep=n_rep
)

summary = summarize_results(results, mu)
for k, v in summary.items():
    print(f"{k}: {v:.4f}")
```

```python
# Sampling distribution plot
plot_sampling_distribution(results, mu)
plt.show()
```

```python
# Confidence intervals plot
plot_confidence_intervals(results, mu)
plt.show()
```

```python
# LLN convergence experiment
ns = [10, 30, 100, 300, 1000, 3000]
convergence_experiment(
    ns=ns,
    mu=mu,
    sigma=sigma,
    alpha=alpha,
    n_rep=500
)
plt.show()
```
